### Доказательства в арифметике

Докажите (или найдите контрпримеры) следующие теоремы для целых, рациональных, вещественных и битовых чисел:
```
x + y == y + x

(x + y) + z == x + (y + z)
```
x может быть меньше, больше или равен 0;
x в квадрате всегда положителен.

````python
# Решение

# целые
x = Int("x")
y = Int("y")
z = Int("z")

prove(x + y == y + x) # коммутативность
prove(((x + y) + z) == ((x + (y + z)))) # ассоциативность
prove(x + 0 == x) # существование "0"
prove(1 * x == x) # существование "1"
prove(Or(x > 0, x < 0, x == 0))
prove(x * (y + z) == x * y + x * z)
prove(x ** 2 >= 0)
````

### Задача 3. Задача о существовании противоположного числа.

Для любого целого A существует единственное целое B, такое что A + B = 0.

````python
A = Int('A')
B = Int('B')
Bu = Int('Bu')

prove(ForAll([A],
             Exists([B], 
                    And(A + B == 0,
                        ForAll([Bu], 
                               Implies(A + Bu == 0, B == Bu)
                              )
                       )
                   )
            )
     )
````
